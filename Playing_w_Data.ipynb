```python
import pandas as pd

data = pd.read_csv("Activities.csv")

data.head()
```

```python
#Changing Dates
data["Date"] = pd.to_datetime(data["Date"])
```

```python
#Adding a week day column
data["Weekday"] = data["Date"].dt.day_name()
```

```python
#Adding a week column
start_date = data["Date"].min()

# Create "week groups" relative to the start
data["Week"] = ((data["Date"] - start_date).dt.days // 7) + 1
data
```

```python
data.columns.to_list()
```

```python
#Cleaning Columns Dropping Uncessary ones

data = data[["Date", "Week", "Weekday", "Distance", "Calories", "Time", "Avg HR", "Max HR", "Avg Pace", "Best Pace", "Total Ascent", "Total Descent", "Steps"]]
data
```

```python
#Filling in blank values
data["Total Ascent"] = data["Total Ascent"].str.replace("--", "0").astype(int)
data["Total Descent"] = data["Total Descent"].str.replace("--", "0").astype(int)
# This changes the entire rows to floats too
```

```python
data.dtypes
```

```python
#Need to change Calories, Time, Avg Pace, Best PAce, Steps into usable floats to work with
#Calories First
data["Calories"] = data["Calories"].str.replace(",", "").astype(int)
data["Calories"].mean()
```

```python
#Time
data["Time"] = pd.to_timedelta(data["Time"]).dt.total_seconds() / 60
```

```python
#Pace
#Going to make a function
def pace_converter_func(pace):
    m, s = pace.split(":")
    m = float(m)
    s = float(s)
    s = s*(5/3)
    s = s/100
    final_pace = m+s
    return final_pace
pace_converter_func("19:00")
```

```python
#Changing Pace here
data["Avg Pace"] = data["Avg Pace"].apply(pace_converter_func)
data["Avg Pace"].mean()
```

```python
#Best Pace Change
data["Best Pace"] = data["Best Pace"].apply(pace_converter_func)
data["Best Pace"].mean()
```

```python
#Steps
data["Steps"] = data["Steps"].str.replace(",", "").astype(int)
data["Steps"].mean()
```
