```python
import pandas as pd

data = pd.read_csv("Activities.csv")

data.head()
```

```python
data.columns.to_list()
```

```python
#Cleaning Columns Dropping Uncessary ones

data = data[["Date", "Distance", "Calories", "Time", "Avg HR", "Max HR", "Avg Pace", "Best Pace", "Total Ascent", "Total Descent", "Steps"]]
data
```

```python
#Filling in blank values
data["Total Ascent"] = data["Total Ascent"].str.replace("--", "0").astype(float)
data["Total Descent"] = data["Total Descent"].str.replace("--", "0").astype(float)
# This changes the entire rows to floats too
```

```python
data.dtypes
```

```python
#Need to change Calories, Time, Avg Pace, Best PAce, Steps into usable floats to work with
#Calories First
data["Calories"] = data["Calories"].str.replace(",", "").astype(float)
data["Calories"].mean()
```

```python
#Time
data["Time"] = pd.to_timedelta(data["Time"]).dt.total_seconds() / 60
```

```python
#Pace
#Going to make a function
def pace_converter_func(pace):
    m, s = pace.split(":")
    m = float(m)
    s = float(s)
    s = s*(5/3)
    s = s/100
    final_pace = m+s
    return final_pace
pace_converter_func("19:00")
```

```python
#Changing Pace here
data["Avg Pace"] = data["Avg Pace"].apply(pace_converter_func)
data["Avg Pace"].mean()
```
